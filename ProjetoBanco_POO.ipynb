```python
import random

class contaCorrente:
    def __init__(self, titular, senha):
        self.titular = titular
        self.conta = random.randint(100, 999)
        self.__senha = senha
        self.__saldoC = 0.0

    def get_saldoC(self):
        return self.__saldoC

    def set_saldoC(self, saldoC):
        self.__saldoC = saldoC

    def get_senha(self):
        return self.__senha

    def set_senha(self, senha):
        self.__senha = senha

    def depositar(self, valor):
        self.__saldoC += valor
        print(f"Depósito de R$ {valor:.2f} efetuado com sucesso!")

    def sacar(self, valor):
        if valor <= self.__saldoC:
            self.__saldoC -= valor
            print(f"Saque de R$ {valor:.2f} efetuado com sucesso!")
        else:
            print("Saldo insuficiente para saque.")

    def aplicar(self, valor, conta_poupanca):
        if valor <= self.__saldoC:
            self.__saldoC -= valor
            conta_poupanca.aplicar(valor)
            print(f"Aplicação de R$ {valor:.2f} na conta poupança efetuada com sucesso!")
        else:
            print("Saldo insuficiente para a aplicação.")

class ContaPoupanca:
    def __init__(self):
        self.__saldoP = 0.0

    def get_saldoP(self):
        return self.__saldoP

    def set_saldoP(self, saldoP):
        self.__saldoP = saldoP

    def aplicar(self, valor):
        self.__saldoP += valor
        print(f"R$ {valor:.2f} aplicado na conta poupança!")

    def resgatar(self, valor):
        if valor <= self.__saldoP:
            self.__saldoP -= valor
            print(f"Resgate de R$ {valor:.2f} da conta poupança efetuado com sucesso!")
            return valor
        else:
            print("Saldo insuficiente para o resgate.")
            return 0.0

def criarConta():
    print("Nome do Banco: Banco do Brasil")
    print("Cadastrar Conta")
    titular = input("Digite seu nome: ")
    senha = cadastrarSenha()
    conta_corrente = contaCorrente(titular, senha)
    conta_poupanca = ContaPoupanca()

    print(f"\nConta criada provisoriamente! Número da conta: {conta_corrente.conta}")
    print("Para ativar a conta, realize um depósito inicial de no mínimo R$ 10,00.")

    while True:
        try: #aqui não teve jeito hehe :)
            deposito_inicial = float(input("Valor do depósito inicial: R$ "))
            if deposito_inicial >= 10:
                conta_corrente.depositar(deposito_inicial)
                print("Conta ativada com sucesso!")
                return conta_corrente, conta_poupanca
            else:
                print("O depósito inicial deve ser de no mínimo R$ 10,00.")
        except ValueError:
            print("Por favor, insira um valor válido para o depósito.")

def cadastrarSenha():
    senha = input("Digite uma senha de 4 dígitos: ")
    if senha.isdigit() and len(senha) == 4:
        return senha
    else:
        print("A senha deve conter apenas 4 caracteres númericos!")
        return cadastrarSenha()

def verificacaoDeSenha(senha_cadastrada):
    for i in range(3):
        senha_informada = input("Digite sua senha: ")
        if senha_informada == senha_cadastrada:
            print("Senha correta! Seja bem-vindo.")
            return True
        elif (2 - i) > 0:
            print(f"Senha incorreta! Você tem {2 - i} tentativas restantes.")
        else:
            print("Conta bloqueada! Dirija-se para agência com documento com foto para desbloqueio da senha!")
            return False

def extrato(conta_corrente, conta_poupanca):
    print("\n+------------------------------------------------+")
    print(f"| Titular: {conta_corrente.titular}")
    print(f"| Número da conta: {conta_corrente.conta}")
    print(f"| Saldo da Conta Corrente: R$ {conta_corrente.get_saldoC():.2f}")
    print(f"| Saldo da Conta Poupança: R$ {conta_poupanca.get_saldoP():.2f}")
    print("+------------------------------------------------+")

def main():
    conta_corrente, conta_poupanca = criarConta()
    if verificacaoDeSenha(conta_corrente.get_senha()):
        print("Bem-vindo ao sistema!")
        while True:
            print("\nOperações disponíveis:")
            print("1 - Depositar")
            print("2 - Sacar")
            print("3 - Aplicar na Poupança")
            print("4 - Resgatar da Poupança")
            print("5 - Extrato")
            print("6 - Sair")
            opcao = input("Escolha uma operação: ")

            if opcao == '1':
                valor = float(input("Valor para depósito: R$ "))
                conta_corrente.depositar(valor)
            elif opcao == '2':
                valor = float(input("Valor para saque: R$ "))
                conta_corrente.sacar(valor)
            elif opcao == '3':
                valor = float(input("Valor para aplicar na poupança: R$ "))
                conta_corrente.aplicar(valor, conta_poupanca)
            elif opcao == '4':
                valor = float(input("Valor para resgatar da poupança: R$ "))
                valor_resgatado = conta_poupanca.resgatar(valor)
                if valor_resgatado > 0:
                    conta_corrente.depositar(valor_resgatado)
            elif opcao == '5':
                extrato(conta_corrente, conta_poupanca)
            elif opcao == '6':
                print("Saindo do sistema. Até logo!")
                break

    else:
        print("Acesso negado.")

if __name__ == "__main__":
    main()
```
